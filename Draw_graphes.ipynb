# Histogramme of text length

```python
raw_context = []
with open("datasets/news_less_category.txt", "r") as f:
    for line in f:
        temp = line.split('__label__')
        raw_context.append(temp[0])
        
# We tokenize each text 
tokenized_context = [sentence.split() for sentence in raw_context]

# We count length of each text
lens = []
for i in range(len(tokenized_context)):
    lens.append(len(tokenized_context[i]))
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(121)
plt.hist(lens, bins=100, range=(0, 10000))
#plt.yscale('log')
plt.xlabel("Longueur de texte", fontsize=14)
plt.ylabel('Nombre de texte', fontsize=14)
plt.title("Histogramme de la longueur de texte", fontsize=16)

plt.subplot(122)
plt.hist(lens, bins=100, range=(0, 10000), cumulative=True)
plt.xlabel("Longueur de texte", fontsize=14)
plt.ylabel('Nombre de texte cumulé', fontsize=14)
plt.title("Histogramme cumulé de la longueur de texte", fontsize=16)
plt.savefig('histo_length.png', dpi=300, bbox_inches='tight')
plt.show()
#plt.yscale('log')
#print(len(lens))
```
